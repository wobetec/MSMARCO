```python
import ir_datasets
import random
import pickle
from tqdm import tqdm
import numpy as np
from pickle import load
import json
```

```python
with open('subset_msmarco_train_0/subset_msmarco_train_0.1_9.pkl','rb') as f:
    data = load(f)
```

```python
with open('docs.jsonl', 'w') as f:
    for _, doc in data['docs'].items():
        json_line = {'id': doc.doc_id, 'contents': doc.text}
        f.write(json.dumps(json_line) + '\n')
```

```python
with open('queries.jsonl', 'w') as f:
    for _, query in data['queries'].items():
        json_line = {'id': query.query_id, 'contents': query.text}
        f.write(json.dumps(json_line) + '\n')
```

```python
with open('qrels.tsv', 'w') as f:
    for qrel in data['qrels']:
        # Extract fields from the TrecQrel object
        query_id = qrel.query_id
        doc_id = qrel.doc_id
        relevance = qrel.relevance
        # Write in the required format: <query_id> 0 <doc_id> <relevance>
        f.write(f"{query_id} 0 {doc_id} {relevance}\n")
```
